```python
# Analise exploratória dos dados
# Dataset: titanic
# Colunas:
#     PassengerId: Id unico do passageiro
#     Survived: 0 não, 1 sim
#     Pclass: 1 - primeira, 2 - segunda, 3 - terceira
#     Name: nome do passageiro
#     Sex: genero do passageiro
#     Age: idade (em anos)
#     SibSp: irmãos/conjuges a bordo
#     Parch: numero de pais / filhos a bordo
#     Ticket: n. do bilhete
#     Fare: valor do tiquet
#     Cabin: n. da cabine
#     Embarked: C - Cherbourg, Q - Queenstown, S - Southampton
```

```python
import pandas as pd
dados = pd.read_csv('titanic_train.csv')
display(dados)
```

```python
# verificando o n. de valores únicos
dados.nunique()
```

```python
# visualização de forma gráfica
import matplotlib.pyplot as plt
```

```python
x = dados.Fare

fig, ax = plt.subplots()

ax.hist(x, bins = 20, linewidth=0.5, edgecolor = 'white')

plt.show()
```

```python
dados['Embarked'].value_counts()
```

```python
import numpy as np

plt.style.use('_mpl-gallery-nogrid')
x = dados.Embarked.value_counts()
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))

# plot
fig, ax = plt.subplots()
ax.pie(x, colors=colors, radius=3, center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True, labels=["S", "C", "Q"])

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
